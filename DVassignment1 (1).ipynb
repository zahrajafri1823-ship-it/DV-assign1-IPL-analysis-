# DV Assignment 1 (IPL Analysis) 


```python
import pandas as pd
import matplotlib.pyplot as plt

deliveries = pd.read_csv('deliveries.csv')
matches = pd.read_csv('matches.csv')
```

## Question 1) Match won by the maximum margin of runs.

### Code:

```python
max_runs = matches['win_by_runs'].max()
matches.loc[matches['win_by_runs'].idxmax()]
```

### Graph:

```python
plt.figure(figsize=(15,6))
plt.plot(matches.index[ : :15], matches['win_by_runs'][ : :15], color = 'orange', label = 'win by runs', linestyle = '-',marker='o')
plt.title('Win By Runs - All Matches')
plt.xlabel('Match Index')
plt.ylabel('Runs')
plt.legend()
plt.show()
```

### Insights:

#### 1. This is the highest winning margin by runs.
#### 2. The printed row shows complete match details.

## Question 5) Matches where D/L method was and wasn't applied.

### Code:

```python
dl_counts = matches['dl_applied'].value_counts()
dl_applied = dl_counts.get(1, 0)
dl_not_applied = dl_counts.get(0, 0)
print("\nMatches where D/L Applied:", dl_applied)
print("Matches where D/L Not Applied:", dl_not_applied)
```

### Graph:

```python
plt.figure(figsize=(6,4))
plt.bar(['DL Applied','DL Not Applied'],[dl_applied, dl_not_applied], color=['red','green'])
plt.title('DL Method Usage in Matches')
plt.ylabel('Number of Matches')
plt.show()
```

### Insight:

#### 1. DL method was applied in very few matches compared to normal matches.
#### 2. Most IPL matches were completed without rain interruption.

## Question 10) Does winning toss mean winning match?

### Code:

```python
toss_win_match = matches[matches['toss_winner'] == matches['winner']]
result = toss_win_match.shape[0]
total = matches.shape[0]
print("Matches where Toss Winner won Match:", result)
print("Total Matches:", total)
```

### Graph:

```python
plt.figure(figsize=(6,4))
plt.bar(['Won After Toss','Lost After Toss'],[result, total - result],color=['pink','skyblue'])
plt.title('Toss Impact on Match Result')
plt.ylabel('Matches')
plt.show()
```

### Insights:

#### 1. Toss does not guarantee match win.
#### 2. Many matches are won by toss-losing teams.
